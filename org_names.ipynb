### Erstellen einer sqlite Datenbank aus Excel

```python
import sqlite3
import pandas as pd

database = sqlite3.connect('data.db')
database.execute("DROP TABLE IF EXISTS org")
database.execute("CREATE TABLE org(name varchar, city varchar)")
df = pd.read_excel("org.xlsx")
for idx in df.index:
    row = df.loc[idx]
    name = row["name"]
    city = row["stadt"]
    database.execute(f'INSERT INTO org(name, city) VALUES(?, ?)', [name, city])
```

### Beispielhaftes Laden der Daten

```python
rows = database.execute("select * from org").fetchall()
```
